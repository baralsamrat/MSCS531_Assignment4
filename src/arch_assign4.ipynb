```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Fabricate dataset simulating attempts until a match is found
# Simulating 100 trials, where the number of attempts to find a match ranges between 1 and 20
np.random.seed(42)  # For reproducibility
num_trials = 100
attempts_list = np.random.randint(1, 20, size=num_trials)

# Create a DataFrame to store attempts data
df = pd.DataFrame({"Trial": range(1, num_trials + 1), "Attempts": attempts_list})

# Calculate basic statistics and display
stats = df["Attempts"].describe()
print("Summary Statistics:\n", stats)

# Plotting histogram of attempts distribution
plt.figure(figsize=(10, 6))
plt.hist(df["Attempts"], bins=15, edgecolor="black")
plt.title("Distribution of Attempts to Find a Match")
plt.xlabel("Number of Attempts")
plt.ylabel("Frequency")
plt.show()

# Plotting attempts per trial
plt.figure(figsize=(10, 6))
plt.plot(df["Trial"], df["Attempts"], marker="o", linestyle="-")
plt.title("Attempts per Trial")
plt.xlabel("Trial Number")
plt.ylabel("Attempts")
plt.show()
```
